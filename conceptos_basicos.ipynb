```python
import matplotlib.pyplot as plt # Import the matplotlib.pyplot library for plotting
import numpy as np # Import the numpy library for numerical operations

# Generate 1000 data points with correlation
num_points = 1000 # Define the number of data points
advertising = 50 + 70 * np.random.rand(num_points) # Generate random data for advertising
# Generate sales data with a linear relationship to advertising and some noise
sales = 100 + 1.5 * advertising + 50 * np.random.rand(num_points) # Generate sales data with correlation to advertising

# Create the scatter plot
plt.figure(figsize=(8, 6)) # Create a figure and set its size
plt.scatter(advertising, sales, alpha=0.5) # Create a scatter plot with advertising on the x-axis and sales on the y-axis
plt.xlabel('Publicidad') # Set the label for the x-axis
plt.ylabel('Ventas') # Set the label for the y-axis
plt.title('Grfico de Dispersin: Ventas vs. Publicidad (1000 Puntos con Correlacin)') # Set the title of the plot
plt.grid(True) # Add a grid to the plot
plt.show() # Display the plot
```

```python
import math

def area_circulo(radio):
  """Calcula el rea de un crculo dado su radio."""
  return math.pi * radio**2

# Ejemplo de uso:
radio_ejemplo = 5
area = area_circulo(radio_ejemplo)
print(f"El rea de un crculo con radio {radio_ejemplo} es: {area}")
```

```python
def celsius_a_fahrenheit(celsius):
  """Convierte grados Celsius a Fahrenheit."""
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit

# Ejemplo de uso:
grados_celsius = 25
grados_fahrenheit = celsius_a_fahrenheit(grados_celsius)
print(f"{grados_celsius} grados Celsius son {grados_fahrenheit} grados Fahrenheit.")
```
